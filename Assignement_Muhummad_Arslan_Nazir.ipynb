# Q1. Arithmetic and Assignment Operators

```python
# Create two variables
a = 10
b = 3

# Arithmetic operations
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)

# Assignment operators
a += 2
print("a after += 2 is", a)

a -= 1
print("a after -= 1 is", a)

a *= 3
print("a after *= 3 is", a)

a /= 2
print("a after /= 2 is", a)
```

# Q2. Comparison and Logical Operators

```python
# Take inputs from user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Comparison operations
print("num1 == num2:", num1 == num2)
print("num1 != num2:", num1 != num2)
print("num1 > num2:", num1 > num2)
print("num1 < num2:", num1 < num2)
print("num1 >= num2:", num1 >= num2)
print("num1 <= num2:", num1 <= num2)

# Logical operations
print("Both are greater than 0:", num1 > 0 and num2 > 0)
print("At least one is greater than 0:", num1 > 0 or num2 > 0)
print("Not equal:", not num1 == num2)
```

# Q3. Bitwise Operators

```python
# Take two integers
x = int(input("Enter first integer: "))
y = int(input("Enter second integer: "))

# Bitwise operations
print("AND:", x & y)
print("OR:", x | y)
print("XOR:", x ^ y)
print("NOT x:", ~x)
print("Left Shift x:", x << 1)
print("Right Shift x:", x >> 1)
```

# Q4. Identity and Membership Operators

```python
# Create a list
my_list = [1, 2, 3, 4, 5]

# Variables to test identity
list2 = my_list
list3 = [1, 2, 3, 4, 5]

# Identity operators
print("list2 is my_list:", list2 is my_list)
print("list3 is not my_list:", list3 is not my_list)

# Membership operators
print("Is 3 in my_list:", 3 in my_list)
print("Is 10 not in my_list:", 10 not in my_list)
```

# Q5. String Operations

```python
# Create a string
text = "I am learning python today."

# Length
print("Length:", len(text))

# Upper and lower case
print("Upper:", text.upper())
print("Lower:", text.lower())

# Find a word
print("Is python in text:", "python" in text)

# Replace a word
new_text = text.replace("python", "coding")
print("Replaced:", new_text)

# Slice a portion (getting "learning")
print("Sliced:", text[5:13])

# Reverse the string
print("Reversed:", text[::-1])
```
